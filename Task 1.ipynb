# Q1 : Create Your own filter and reduce function same as real filter and reduce function?

```python
#Reduce and Filter Functions
from functools import reduce
```

```python
numbers = [1,2,3,4,5]

def multiply(a,b):
  return a*b

reduce(multiply,numbers)
```

```python
#my own reduce function

def my_reduce(func,iterable):
  iterator = iter(iterable)
  result = next(iterator)

  for i in iterator:
     result = func(result,i)

  return result

numbers = [1,2,3,4,5]

def multiply(a,b):
  return a*b
```

```python
my_reduce(multiply,numbers)
```

# Q2: Move Files from one folder to another folder. By using Python and create empty files using python.

```python
import shutil
import os
#move Files
def mov_files(og,tg):
    files = os.listdir(og)
    for f in files:
        shutil.move(og+f,tg)
```

# Q6: Scrape mobile name, mobile price and reviews from first 20 page of flipkart. 


```python
import requests
from bs4 import BeautifulSoup as bs

url = "https://www.flipkart.com/search?q=mobiles"

def scrape_page_data(url):
    url_data = requests.get(url)
    soup = bs(url_data.text, 'html.parser')
    
    mobiles = []
    
    for card in soup.find_all('div',{'class':'_13oc-S'}):
        name3 = card.find('div',{'class':'_4rR01T'})
        price = card.find('div',{'class':'_30jeq3 _1_WHN1'})
        review = card.find('div',{'class','_3LWZlK'})
        mobiles.append({'name': name3.text, 'price': price.text, 'reviews': review.text})
    
    return mobiles


def scrape_data(num_pages):
    base_url = 'https://www.flipkart.com/search?q=mobiles'
    all_mobiles = []

    for page in range(1, num_pages + 1):
        url = f'{base_url}&page={page}'
        mobiles = scrape_page_data(url)
        all_mobiles.extend(mobiles)

    return all_mobiles
```

```python
result = scrape_data(20)

for mobile in result:
    print('Name:', mobile['name'])
    print('Price:', mobile['price'])
    print('Reviews:', mobile['reviews'])
    print('---')
```
